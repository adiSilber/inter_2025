```python
def sample_to_text(sample):
    return ''.join(sample.question_formatted_contents_tokenized + sample.upto_injection_tokens + sample.injection_tokenized + sample.after_injection_tokens)
```

```python
len_of_injection = len(sample0.injection_tokenized)
len_to_injection = len(sample0.question_formatted_contents_tokenized + sample0.upto_injection_tokens)

# Stack all layers directly
nice_matrix = torch.cat([
    sample0.activations_after_injection["layer_1_attention"][i][0, 0, :, len_to_injection:len_to_injection + len_of_injection]
    for i in range(30)
]).float()

plt.figure(figsize=(6, 4))
sns.heatmap(nice_matrix.T, cmap='coolwarm')
plt.title('Head 1, Layer 0 attention')
plt.xlabel('Tokens after injection index')
plt.ylabel('Tokens of injection index')
plt.show()
```

```python
len_of_injection = len(sample0.injection_tokenized)
len_to_injection = len(sample0.question_formatted_contents_tokenized + sample0.upto_injection_tokens)

number_of_layers = 28
fig, axes = plt.subplots(number_of_layers, 1, figsize=(20, 100))  # Adjust size as needed

tokens_after_injection = 30

for layer in tqdm(range(number_of_layers)):
    s = torch.zeros(tokens_after_injection, len_of_injection)
    for head in range(28):
        head_attention_matrix = torch.cat([
            sample0.activations_after_injection[f"layer_{layer}_attention"][i][0, head, :, len_to_injection:len_to_injection + len_of_injection]
            for i in range(tokens_after_injection)
        ]).float()
        s += head_attention_matrix
        
    sns.heatmap(s.T, cmap='coolwarm', ax=axes[layer], cbar=False)
    axes[layer].set_title(f'L{layer} Sum of head attentions', fontsize=6)
    axes[layer].tick_params(labelsize=4)

plt.xlabel('Tokens after injection index')
plt.ylabel('Tokens of injection index')
plt.show()
```
